```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

Data Loading and Basic exploration 

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
movies.head()
```

```python
credits.head()
```

```python
movies=movies.merge(credits,on="title")
movies.head()
```

```python
movies.info()
```

```python
movies=movies[['genres','movie_id','cast','overview','title','keywords','crew']]
movies.head()
```

Data Cleaning

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert1(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert1)
movies.head()
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

```python
 movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
 movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
 movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
 movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
 movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

```python
movies['tag1']=movies.genres+movies.cast+movies.crew
movies['tag2']=movies.overview+movies.keywords
movies.head()
```

```python
movies=movies[['movie_id','title','tag1','tag2']]
movies.head()
```

```python
movies.tag1=movies.tag1.apply(lambda x:" ".join(x))
movies.tag2=movies.tag2.apply(lambda x:" ".join(x))
```

```python
movies.head()
```

```python
movies.tag1=movies.tag1.apply(lambda x:x.lower())
movies.tag2=movies.tag2.apply(lambda x:x.lower())
movies.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies.tag1=movies.tag1.apply(stem)
movies.tag2=movies.tag2.apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
vector2 = tfidf.fit_transform(movies['tag2'])

```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vector1 = cv.fit_transform(movies['tag1'])
```

```python
from scipy.sparse import hstack
from sklearn.preprocessing import normalize
# Normalize sparse matrices
vector1 = normalize(vector1)
vector2 = normalize(vector2)
vectors=hstack((vector1, vector2))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity
```

Creating the recommending function based on their similarity score 

```python
def recommend(movie):
    if not movies[movies.title == movie].empty:
        movie_index = movies[movies.title == movie].index[0]
    else:
        print("Movie not found")
    movies_list=sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies.title.iloc[i[0]])
```

```python
recommend('Avatar')
```
